```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Function to format Rupiah
def format_rupiah(amount):
    return f"Rp {int(amount):,}".replace(",", ".")

# Function to format percentage
def format_percent(decimal):
    return f"{decimal*100:.0f}%"

# Load datasets
aset_tetap = pd.read_csv('aset_tetap.csv')
kebijakan_fiskal = pd.read_csv('kebijakan_fiskal.csv')
transaksi_keuangan = pd.read_csv('transaksi_keuangan.csv')

# Ensure numeric data types
transaksi_keuangan[['pendapatan', 'beban_operasional', 'penyusutan']] = transaksi_keuangan[['pendapatan', 'beban_operasional', 'penyusutan']].astype(float)
aset_tetap[['nilai_perolehan', 'umur_ekonomis']] = aset_tetap[['nilai_perolehan', 'umur_ekonomis']].astype(float)
kebijakan_fiskal[['tax_rate']] = kebijakan_fiskal[['tax_rate']].astype(float)

# Perintah 1: Simulate Profit/Loss and Taxes
def simulate_profit_loss():
    results = []
    for _, row in transaksi_keuangan.iterrows():
        year = row['tahun']
        pendapatan = row['pendapatan']
        beban_operasional = row['beban_operasional']
        penyusutan = row['penyusutan']
        skenario = row['skenario']

        # Calculate gross profit/loss
        laba_kotor = pendapatan - beban_operasional - penyusutan
        laba_status = 'Laba' if laba_kotor >= 0 else 'Rugi'

        # Get tax rate for the year
        tax_row = kebijakan_fiskal[kebijakan_fiskal['tahun'] == year]
        if not tax_row.empty:
            tax_rate = tax_row['tax_rate'].iloc[0]
        else:
            tax_rate = 0  # Default if no tax policy

        # Tax calculation (no tax on loss)
        pajak_laba = max(laba_kotor * tax_rate, 0)
        pajak_rugi = 0  # No tax on loss

        results.append({
            'Tahun': year,
            'Skenario': skenario,
            'Pendapatan': format_rupiah(pendapatan),
            'Beban Operasional': format_rupiah(beban_operasional),
            'Penyusutan': format_rupiah(penyusutan),
            'Laba Kotor': format_rupiah(laba_kotor),
            'Status': laba_status,
            'Tax Rate': format_percent(tax_rate),
            'Pajak (Laba)': format_rupiah(pajak_laba),
            'Pajak (Rugi)': format_rupiah(pajak_rugi)
        })

    return pd.DataFrame(results)

# Perintah 2: Simulate Depreciation (Straight-Line and Double-Declining Balance)
def simulate_depreciation():
    results = []
    for year in transaksi_keuangan['tahun']:
        tax_row = kebijakan_fiskal[kebijakan_fiskal['tahun'] == year]
        if tax_row.empty:
            continue
        tax_rate = tax_row['tax_rate'].iloc[0]

        # Calculate total depreciation per method
        total_sl = 0
        total_ddb = 0
        for _, asset in aset_tetap.iterrows():
            nilai = asset['nilai_perolehan']
            umur = asset['umur_ekonomis']

            # Straight-Line
            sl_depr = nilai / umur
            total_sl += sl_depr

            # Double-Declining Balance
            ddb_rate = 2 / umur
            book_value = nilai
            ddb_depr = book_value * ddb_rate
            total_ddb += ddb_depr

        # Get financial data
        fin_row = transaksi_keuangan[transaksi_keuangan['tahun'] == year]
        if fin_row.empty:
            continue
        pendapatan = fin_row['pendapatan'].iloc[0]
        beban_operasional = fin_row['beban_operasional'].iloc[0]

        # Calculate taxable income and tax
        laba_sl = pendapatan - beban_operasional - total_sl
        laba_ddb = pendapatan - beban_operasional - total_ddb
        pajak_sl = max(laba_sl * tax_rate, 0)
        pajak_ddb = max(laba_ddb * tax_rate, 0)

        results.append({
            'Tahun': year,
            'Pendapatan': format_rupiah(pendapatan),
            'Beban Operasional': format_rupiah(beban_operasional),
            'Penyusutan Garis Lurus': format_rupiah(total_sl),
            'Penyusutan Saldo Menurun': format_rupiah(total_ddb),
            'Laba Kotor (Garis Lurus)': format_rupiah(laba_sl),
            'Laba Kotor (Saldo Menurun)': format_rupiah(laba_ddb),
            'Tax Rate': format_percent(tax_rate),
            'Pajak (Garis Lurus)': format_rupiah(pajak_sl),
            'Pajak (Saldo Menurun)': format_rupiah(pajak_ddb)
        })

    return pd.DataFrame(results)

# Perintah 3: Simulate Tax Holiday
def simulate_tax_holiday():
    results = []
    for year in transaksi_keuangan['tahun']:
        tax_row = kebijakan_fiskal[kebijakan_fiskal['tahun'] == year]
        if tax_row.empty:
            continue
        tax_rate = tax_row['tax_rate'].iloc[0]
        tax_holiday_awal = tax_row['tax_holiday_awal'].iloc[0]
        tax_holiday_akhir = tax_row['tax_holiday_akhir'].iloc[0]

        # Check if year is within tax holiday
        is_tax_holiday = tax_holiday_awal <= year <= tax_holiday_akhir

        # Get financial data
        fin_row = transaksi_keuangan[transaksi_keuangan['tahun'] == year]
        if fin_row.empty:
            continue
        pendapatan = fin_row['pendapatan'].iloc[0]
        beban_operasional = fin_row['beban_operasional'].iloc[0]
        penyusutan = fin_row['penyusutan'].iloc[0]

        # Calculate gross profit
        laba_kotor = pendapatan - beban_operasional - penyusutan

        # Tax calculations
        pajak_normal = max(laba_kotor * tax_rate, 0)
        pajak_holiday = 0 if is_tax_holiday else pajak_normal

        results.append({
            'Tahun': year,
            'Laba Kotor': format_rupiah(laba_kotor),
            'Tax Rate': format_percent(tax_rate),
            'Tax Holiday': 'Ya' if is_tax_holiday else 'Tidak',
            'Pajak (Tax Holiday)': format_rupiah(pajak_holiday),
            'Pajak (Non-Tax Holiday)': format_rupiah(pajak_normal)
        })

    return pd.DataFrame(results)

# Display results
print("=== Perintah 1: Simulasi Laba/Rugi dan Pajak ===")
profit_loss_df = simulate_profit_loss()
display(profit_loss_df)

print("\n=== Perintah 2: Simulasi Penyusutan dan Pajak ===")
depreciation_df = simulate_depreciation()
display(depreciation_df)

print("\n=== Perintah 3: Simulasi Tax Holiday ===")
tax_holiday_df = simulate_tax_holiday()
display(tax_holiday_df)

# Perintah 4: Visualizations
plt.style.use('seaborn')

# Plot 1: Laba Kotor per Tahun
plt.figure(figsize=(10, 6))
sns.barplot(data=profit_loss_df, x='Tahun', y=profit_loss_df['Laba Kotor'].str.replace('Rp ', '').str.replace('.', '').astype(float), hue='Skenario')
plt.title('Laba Kotor per Tahun berdasarkan Skenario')
plt.ylabel('Laba Kotor (Rp)')
plt.xlabel('Tahun')
plt.legend(title='Skenario')
plt.tight_layout()
plt.show()

# Plot 2: Pajak Laba vs Rugi
plt.figure(figsize=(10, 6))
pajak_laba = profit_loss_df['Pajak (Laba)'].str.replace('Rp ', '').str.replace('.', '').astype(float)
pajak_rugi = profit_loss_df['Pajak (Rugi)'].str.replace('Rp ', '').str.replace('.', '').astype(float)
plt.plot(profit_loss_df['Tahun'], pajak_laba, label='Pajak (Laba)', marker='o')
plt.plot(profit_loss_df['Tahun'], pajak_rugi, label='Pajak (Rugi)', marker='s')
plt.title('Pajak Laba vs Rugi per Tahun')
plt.ylabel('Pajak (Rp)')
plt.xlabel('Tahun')
plt.legend()
plt.tight_layout()
plt.show()

# Plot 3: Penyusutan Garis Lurus vs Saldo Menurun
plt.figure(figsize=(10, 6))
sl_depr = depreciation_df['Penyusutan Garis Lurus'].str.replace('Rp ', '').str.replace('.', '').astype(float)
ddb_depr = depreciation_df['Penyusutan Saldo Menurun'].str.replace('Rp ', '').str.replace('.', '').astype(float)
plt.plot(depreciation_df['Tahun'], sl_depr, label='Garis Lurus', marker='o')
plt.plot(depreciation_df['Tahun'], ddb_depr, label='Saldo Menurun Ganda', marker='s')
plt.title('Penyusutan: Garis Lurus vs Saldo Menurun Ganda')
plt.ylabel('Penyusutan (Rp)')
plt.xlabel('Tahun')
plt.legend()
plt.tight_layout()
plt.show()

# Plot 4: Pajak dengan dan tanpa Tax Holiday
plt.figure(figsize=(10, 6))
pajak_holiday = tax_holiday_df['Pajak (Tax Holiday)'].str.replace('Rp ', '').str.replace('.', '').astype(float)
pajak_non_holiday = tax_holiday_df['Pajak (Non-Tax Holiday)'].str.replace('Rp ', '').str.replace('.', '').astype(float)
plt.plot(tax_holiday_df['Tahun'], pajak_holiday, label='Tax Holiday', marker='o')
plt.plot(tax_holiday_df['Tahun'], pajak_non_holiday, label='Non-Tax Holiday', marker='s')
plt.title('Pajak dengan dan tanpa Tax Holiday')
plt.ylabel('Pajak (Rp)')
plt.xlabel('Tahun')
plt.legend()
plt.tight_layout()
plt.show()
```
